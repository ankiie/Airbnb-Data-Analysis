**Data Overview**

This is a dataset with 1 million rows contains detailed information about Airbnb property listings, which can be used to analyze market trends, pricing strategies and location-based performance.

**Column Description**

- id – Unique identifier for each listing.

- name – Title or description of the listing.

- host_id and host_name – Host identification details.

- neighbourhood_group and neighbourhood – Location details.

- latitude & longitude – Geographical coordinates.

- room_type – Type of accommodation (e.g., Entire home/apt, Private room).

- price – Cost per night of stay.

- minimum_nights – Minimum number of nights required for booking.

- number_of_reviews – Total number of reviews received.

- last_review – Date of the most recent review.

- reviews_per_month – Average number of reviews per month.

- availability_365 – Number of available booking days in a year.

**Checking Missing Values:**

```python
df.isnull().sum()
```

**Handling Missing Values**

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset=['NAME', 'host name'],inplace =True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["license", "house_rules"],errors='ignore')
```

```python
df.head()
```

```python
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

**Visualization**

**What is the distribution of listing prices?**

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'],bins = 50,kde=True,color='red')
plt.title("Distribution of listing price")
plt.xlabel("price $")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings in any specific price range.The KDE line helps visualize this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices.

**How are different room types distributed?**

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x="room type",data = df, color = "hotpink")
plt.title("Room Type Distribution")
plt.xlabel("room type")
plt
plt.show()
```

---
**How are listings distributed across different neighbourhoods?**

```python
plt.figure(figsize = (12,8))
sns.countplot(y='neighbourhood group', data=df,color="lightgreen", order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

---
**What is the relationship between price and room type?**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price', hue='room type', data=df, palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

---
**How has the number of reviews change over with time?**

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
